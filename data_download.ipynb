```python
from datasets import load_dataset
from PIL import Image
import os


# Specify where to save the images
save_dir = r"C:\Users\crazycyt\Desktop\1470\dataset2" 
os.makedirs(save_dir, exist_ok=True)

# Start index (assuming you already saved the first 22600 images)
start_idx = 22600

# Get the total number of images
total_images = len(ds["train"])

# Loop through and resume saving from where you left off
for idx in range(start_idx, total_images):
    item = ds["train"][idx]
    image = item["image"]
    image_path = os.path.join(save_dir, f"image_{idx}.png")
    image.save(image_path)

    if idx % 100 == 0:
        print(f"Saved {idx} / {total_images} images")

print("Done saving remaining images!")
```
